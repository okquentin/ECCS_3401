```python
# Import Required Packages
import pandas as pd
import math
import csv
import requests
from statistics import mean
```

```python
# We're going to pull data on penguins! 
data_url =  "https://raw.githubusercontent.com/mcnakhaee/palmerpenguins/refs/heads/master/palmerpenguins/data/penguins.csv"
# Pandas is a library that has holds data in a tabular format
penguins = pd.read_csv(data_url)
penguins
```

```python
# List All Species
species_list = penguins["species"].unique()
```

```python
# Loop through each species, calculate/display average weights, find largest
def find_largest_weight():
    avgWeights = {}
    for species in species_list:
        speciesWeights = penguins.loc[penguins["species"] == species, "body_mass_g"].dropna()

        totalWeight = 0
        count = 0

        for weight in speciesWeights:
            totalWeight += weight
            count += 1

        avg = totalWeight / count if count > 0 else None
        avgWeights[species] = avg
        print(f"Average Weight (g) for {species}: {avg:.2f}")
    largestWeightSpecies = max(avgWeights, key=avgWeights.get)
    print(f"========================================")
    print(f"Largest Average Weight (g): {avgWeights[largestWeightSpecies]:.2f}")
    print(f"========================================")
```

```python
# Loop through each species, calculate/display average lengths, find smallest
def find_smallest_length():
    avgLengths = {}
    for species in species_list:
        speciesLength = penguins.loc[penguins["species"] == species, "flipper_length_mm"].dropna()

        totalLength = 0
        count = 0

        for length in speciesLength:
            totalLength += length
            count += 1

        avg = totalLength / count if count > 0 else None
        avgLengths[species] = avg
        print(f"Average Length (mm) for {species}: {avg:.2f}")

    smallestLengthSpecies = min(avgLengths, key=avgLengths.get)
    print(f"========================================")
    print(f"Smallest Average Bill Length (mm): {avgLengths[smallestLengthSpecies]:.2f}")
    print(f"========================================")
```

```python
find_largest_weight()
find_smallest_length()
```
